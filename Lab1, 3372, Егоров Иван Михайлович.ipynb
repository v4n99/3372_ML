```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Ссылка на датасет - https://www.kaggle.com/datasets/sujaykapadnis/spotify-songs?resource=download
#Предметная область данного датасета — музыкальная индустрия и стриминговые сервисы.
#Данные за эту неделю получены из Spotify через пакет spotifyr.
#Люди создали этот пакет, чтобы упростить получение как собственных данных, так и общих метаданных о песнях из API Spotify.
#Характер данных - реальные данные, полученные из Spotify.
```

```python
data = pd.read_csv('spotify_songs.csv')
```

```python
#Описание атрибутов датасета:
#Переменная	Тип данных	Описание
#track_id	Строка	Уникальный идентификатор трека
#track_name	Строка	Название трека
#track_artist	Строка	Исполнитель трека
#track_popularity	Число с плавающей точкой	Популярность трека (от 0 до 100), где более высокое значение лучше
#track_album_id	Строка	Уникальный идентификатор альбома
#track_album_name	Строка	Название альбома
#track_album_release_date	Строка	Дата выхода альбома
#playlist_name	Строка	Название плейлиста
#playlist_id	Строка	Идентификатор плейлиста
#playlist_genre	Строка	Жанр плейлиста
#playlist_subgenre	Строка	Поджанр плейлиста
#danceability	Число с плавающей точкой	Показатель пригодности трека для танцев на основе темпа, стабильности ритма, силы бита и общей регулярности. 0.0 — наименее танцевальный, 1.0 — наиболее танцевальный
#energy	Число с плавающей точкой	Мера интенсивности и активности трека (от 0.0 до 1.0). Энергичные треки обычно быстрые, громкие и шумные
#key	Число с плавающей точкой	Основная тональность трека. Целые числа соответствуют музыкальным тонам (0 = До, 1 = До-диез/Ре-бемоль и т.д.). Если тональность не определена, значение -1
#loudness	Число с плавающей точкой	Общая громкость трека в децибелах (дБ). Типичные значения от -60 до 0 дБ
#mode	Число с плавающей точкой	Модальность трека: 1 — мажорная тональность, 0 — минорная
#speechiness	Число с плавающей точкой	Показатель наличия речи в треке. Значения выше 0.66 указывают на чисто речевой контент, между 0.33 и 0.66 — на сочетание музыки и речи, ниже 0.33 — на музыкальный контент
#acousticness	Число с плавающей точкой	Уверенность в том, что трек является акустическим (от 0.0 до 1.0)
#instrumentalness	Число с плавающей точкой	Предсказание отсутствия вокала в треке. Значения выше 0.5 указывают на инструментальный трек
#liveness	Число с плавающей точкой	Показатель присутствия аудитории в записи. Значения выше 0.8 указывают на высокую вероятность живого выступления
#valence	Число с плавающей точкой	Показатель позитивности трека (от 0.0 до 1.0). Высокие значения соответствуют позитивным трекам (радостным, весёлым), низкие — негативным (грустным, гневным)
#tempo	Число с плавающей точкой	Оценочный темп трека в ударах в минуту (BPM)
#duration_ms	Число с плавающей точкой	Длительность трека в миллисекундах
```

```python
data.head()
```

```python
# Создаем список числовых столбцов для анализа
numeric_columns = [
    'track_popularity',
    'danceability',
    'energy',
    'key',
    'loudness',
    'mode',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo',
    'duration_ms'
]
```

```python
#Найдем среднее значение всех числовых столбцов
data[numeric_columns].mean()
```

```python
#Найдем СКО для всех числовых столбцов
data[numeric_columns].std()
```

```python
# Гистограммы
plt.figure(figsize=(20, 30))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(7, 2, i)
    sns.histplot(data[column].dropna(), kde=True)
    plt.title(f'Распределение {column}')
    plt.xlabel(column)
    plt.ylabel('Частота')
plt.tight_layout()
plt.show()
```

```python
# Ящики с усами для выявления выбросов
# Выбросы отображаются как отдельные точки за пределами “усов”
plt.figure(figsize=(20, 30))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(7, 2, i)  # 7 строк, 2 столбца
    sns.boxplot(y=data[column])
    plt.title(f'Выбросы в {column}', fontsize=14)
    plt.ylabel(column, fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Анализ пропусков
missing_values = data.isnull().sum() / len(data) * 100
print("Процент пропусков в каждом столбце:")
print(missing_values)
```

```python
# Расчет корреляционной матрицы
corr_matrix = data[numeric_columns].corr()
```

```python
# Визуализация корреляционной матрицы
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Корреляционная матрица')
plt.show()

#Анализ:
#Energy и loudness имеет наивысший коэффициент корреляции - 0.68
#Danceability и valence имеет умеренную положительную связь, коэффициент равен 0.33.
#Также energy имеет связь со столбцами liveness, valence, tempo (0.16, 0.15, 0.15 соответственно)
#Наименьший коэффициент корреляции имеет energy c acousticness - (-0.54).
#Следующий с конца идет loudness и acousticness - (-0.36).
#Популярность трека (track_popularity) не имеет значимой корреляции с другими атрибутами, так как коэффициенты близки к 0.
#Режим (mode) и тональность (key) также не имеют значимых связей с другими атрибутами.
#Длительность (duration_ms) слабо коррелирует с другими атрибутами, коэффициенты близки к 0.
```

```python
# Создаем отдельные графики
plt.figure(figsize=(12, 16))

# График 1: Danceability vs Energy
# Наблюдается положительная корреляция между танцевальностью и энергичностью. Это означает, что треки с более высокой танцевальностью, как правило, также имеют более высокую энергичность. Однако корреляция не является строгой, так как точки не образуют четкой линейной зависимости.
plt.subplot(3, 2, 1)
sns.scatterplot(x='danceability', y='energy', data=data, alpha=0.6)
plt.title('Танцевальность vs Энергичность')
plt.xlabel('Танцевальность')
plt.ylabel('Энергичность')
```

```python
# График 2: Loudness vs Energy
plt.figure(figsize=(12, 16))

# Наблюдается слабая отрицательная корреляция между громкостью и энергичностью. Это означает, что с увеличением громкости (в сторону меньших отрицательных значений) энергичность, как правило, уменьшается. Однако корреляция не является строгой, так как точки не образуют четкой линейной зависимости.

plt.subplot(3, 2, 2)
sns.scatterplot(x='loudness', y='energy', data=data, alpha=0.6)
plt.title('Громкость vs Энергичность')
plt.xlabel('Громкость')
plt.ylabel('Энергичность')
```

```python
# График 3: Valence vs Danceability
plt.figure(figsize=(12, 16))

# Наблюдается положительная корреляция между эмоциональностью и танцевальностью. Это означает, что треки с более высокой эмоциональностью, как правило, также имеют более высокую танцевальность. Однако корреляция не является строгой, так как точки не образуют четкой линейной зависимости.

plt.subplot(3, 2, 3)
sns.scatterplot(x='valence', y='danceability', data=data, alpha=0.6)
plt.title('Эмоциональность vs Танцевальность')
plt.xlabel('Эмоциональность')
plt.ylabel('Танцевальность')
```

```python
# График 4: Tempo vs Loudness
plt.figure(figsize=(12, 16))

# Наблюдается слабая положительная корреляция между темпом и громкостью. Это означает, что с увеличением темпа громкость, как правило, также увеличивается. Однако корреляция не является строгой, так как точки не образуют четкой линейной зависимости.

plt.subplot(3, 2, 4)
sns.scatterplot(x='tempo', y='loudness', data=data, alpha=0.6)
plt.title('Темп vs Громкость')
plt.xlabel('Темп')
plt.ylabel('Громкость')
```

```python
# График 5: Acousticness vs Instrumentalness
plt.figure(figsize=(12, 16))

# Наблюдается слабая отрицательная корреляция между акустичностью и инструментальностью. Это означает, что с увеличением акустичности инструментальность, как правило, уменьшается. Однако корреляция не является строгой, так как точки не образуют четкой линейной зависимости.



plt.subplot(3, 2, 5)
sns.scatterplot(x='acousticness', y='instrumentalness', data=data, alpha=0.6)
plt.title('Акустичность vs Инструментальность')
plt.xlabel('Акустичность')
plt.ylabel('Инструментальность')
```
